# EY Frog Challenge Kaggle Runner

Two-code-cell workflow: leave the defaults alone and run both cells. The default stage is `gpu`, which does the full run on available GPUs and writes the final submission.

```python
from pathlib import Path

GITHUB_REPO = 'AstralJugs69/EY-bio'
GITHUB_REF = 'main'
STAGE = 'gpu'  # gpu | baseline | feature | finalize | submissions

REPO_DIR = Path('/kaggle/working/ey-frog-repo')
ARTIFACT_ROOT = Path('/kaggle/working/artifacts')
RUN_LABEL = None  # Example: 'gpu_try_01'. Leave as None for a new timestamped run folder. For STAGE='submissions', set this to an existing run label.
TPU_ARTIFACT_DIR = None

BATCH_SIZE = 256
EPOCHS = 100
PATIENCE = 12
```

```python
import os
import subprocess
import sys
import urllib.request

BOOTSTRAP_PATH = Path('/kaggle/working/kaggle_bootstrap.py')
BOOTSTRAP_URL = f'https://raw.githubusercontent.com/{GITHUB_REPO}/{GITHUB_REF}/kaggle_bootstrap.py'

token = None
try:
    from kaggle_secrets import UserSecretsClient
    token = UserSecretsClient().get_secret('GITHUB_TOKEN')
except Exception:
    token = os.getenv('GITHUB_TOKEN')

request = urllib.request.Request(BOOTSTRAP_URL)
if token:
    request.add_header('Authorization', f'token {token}')
with urllib.request.urlopen(request) as response:
    BOOTSTRAP_PATH.write_bytes(response.read())

command = [
    sys.executable,
    str(BOOTSTRAP_PATH),
    '--github-repo', GITHUB_REPO,
    '--github-ref', GITHUB_REF,
    '--stage', STAGE,
    '--repo-dir', str(REPO_DIR),
    '--artifact-root', str(ARTIFACT_ROOT),
    '--batch-size', str(BATCH_SIZE),
    '--epochs', str(EPOCHS),
    '--patience', str(PATIENCE),
]
if RUN_LABEL is not None:
    command.extend(['--run-label', RUN_LABEL])
if TPU_ARTIFACT_DIR is not None:
    command.extend(['--tpu-artifact-dir', str(TPU_ARTIFACT_DIR)])
subprocess.run(command, check=True)
```
